## Handle Missing Data

```python
# Import missing data csv (upload car-sales-missing-data.csv to Colab)
car_sales_missing = pd.read_csv("car-sales-missing-data.csv")
car_sales_missing
```

```python
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
```

```python
# It didn't update the dataframe, we have to reassign
car_sales_missing
```

```python
# Reassign to update DataFrame
car_sales_missing["Odometer"] = car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
car_sales_missing
```

```python
# We can also use inplace=True
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"]. mean,
                                     inplace=True)
car_sales_missing
```

```python
# Delete rows with null values in any column
car_sales_missing.dropna(inplace=True)
```

## Create data from existing data

```python
# Column from series
seats_column = pd.Series([5, 5, 5, 5, 5])

# New column called seats
car_sales["Seats"] = seats_column = seats_column
car_sales
```

```python
car_sales["Seats"].fillna(5, inplace=True)
```

```python
car_sales
```

```python
# Column from list (should have the same length as existing dataframe)
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 4.7, 7.6, 8.7, 3.0, 4.5]
car_sales["Fuel per 100KM"] = fuel_economy
car_sales
```

```python
car_sales["Total fuel used"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100KM"]
```

```python
car_sales
```

```python
# Create a column from a single value
car_sales["Number of wheels"] = 4
car_sales
```

```python
car_sales["Passed road safety"] = True
car_sales.dtypes
```

```python
car_sales
```

```python
car_sales = car_sales.drop("Total fuel used", axis=1) #or use inplace=True
```

```python
# shuffle the data (if frac = 0.5, it shufles 50% of the data)
car_sales_shuffled = car_sales.sample(frac=1)
```
