What we've created so far:  
* Scalar: a single number
* Vector: a number with direction (e.g. wind speed and direction)
* Matrix: a 2-dimensional array of numbers
* Tensor: an n-dimensional array of numbers (where n can be any number, a 0-dimensional tensor is a scalar, a 1-dimensional tensor is a vector)

### Creating random tensors

Random tensors are tensors of some arbitrary size which contain random numbers

```python
# Create two random (but the same) tensors
random_1 = tf.random.Generator.from_seed(42) # set seed for reproductibility
random_1 = random_1.normal(shape=(3,2))
random_2 = tf.random.Generator.from_seed(42)
random_2 = random_2.normal(shape=(3,2))

# Are they equal?
random_1, random_2, random_1 == random_2
```

```python
# Other method of generating random tensors
tf.random.normal(shape=[3, 2], seed = 42)
```

### Matrix Multiplication

In machine learning, matrix multiplication is one of the most common tensor operations

There are two rules our tensors (or matrices) need to fulfill if we're going to matrix multiply them:

1. The inner dimensions must match
2. The resulting matrix has the shape of the outer dimensions

```python
# Matrix multiplication in tensorflow
print(tensor)
tf.matmul(tensor, tensor)
```

```python
tensor * tensor
```

```python
# Matrix multiplication with python operator "@"
tensor @ tensor
```

```python
# Exercise
tensor_1 = tf.constant([[1, 2, 5],
                        [7, 2, 1],
                        [3, 3, 3]])
tensor_2 = tf.constant([[3, 5],
                        [6, 7],
                        [1, 8]])

tf.matmul(tensor_1, tensor_2)
```

```python
tensor.shape
```

```python
# Create a (3, 2) tensor
X = tf.constant([[1, 2],
                 [3, 4],
                 [5, 6]])
# Create another (3, 2) tensor
Y = tf.constant([[7, 8],
                 [9, 10],
                 [11, 12]])

X, Y
```

```python
# Try to matrix multiply tensors of same shape
tf.matmul(X, Y)
```

```python
Y
```

```python
# Let's change the shape of Y
tf.reshape(Y, shape=(2, 3))
```

```python
X.shape, tf.reshape(Y, shape=(2, 3)).shape
```

```python
# Try to matrix multiply X by reshaped Y
X @ tf.reshape(Y, shape=(2, 3))
```

```python
tf.matmul(X, tf.reshape(Y, shape=(2, 3)))
```

```python
# Try change the shape of X instead of Y
tf.matmul(tf.reshape(X, shape=(2, 3)), Y)
```

```python
# Can do the same with transpose (same shape but different values)
# Transpose flips the axises, reshape fills the new dimensions with the values of the original
X, tf.transpose(X), tf.reshape(X, shape=(2, 3))
```

```python
# Try matrix multiplication with transpose rather than reshape
tf.matmul(tf.transpose(X), Y)
```

**The dot product**

Matrix multiplication is also referred to as the dot product.

You can perform matrix multiplication using:
* `tf.matmul()`
* `tf.tensordot()`
* `@`

```python
X, Y
```

```python
# Perform the dot product on X and Y (requires X or Y to be transposed)
tf.tensordot(tf.transpose(X), Y, axes=1)
```

```python
# Perform matrix multiplication between X and Y (reshaped)
tf.matmul(X, tf.reshape(Y, shape=(2, 3)))
```

```python
# Check the values of Y, reshape Y and transposed Y
print("Normal Y:")
print(Y, "\n") # \n is for newline

print("Y reshaped to (2, 3):")
print(tf.reshape(Y, (2, 3)), "\n")

print("Y transposed:")
print(tf.transpose(Y))
```

Generally, when performing matrix multiplication on two tensors and one of the axes doesn't line up, you will transpose (rather than reshape) one of the tensors to satisfy the matrix multiplication rules

Aggregating tensors = condensing them from multiple values fown to a smaller amount of values

```python
D = tf.constant([-7, -10])
D
```

```python
# Get the absolute values
tf.abs(D)
```

Let0s go through the following forms of aggregation:
* Get the minimum
* Get the maximum
* Get the mean of a tensor
* Get the sum of a tensor

```python
# Create a random tensor with values between 0 and 100 of size 50
E = tf.constant(np.random.randint(0, 100, size=50))
E
```

```python
tf.size(E), E.shape, E.ndim
```

```python
# Find the minimum
tf.reduce_min(E)
```

```python
# Find the maximum
tf.reduce_max(E)
```

```python
# Find the mean
tf.reduce_mean(E)
```

```python
# Find the sum
tf.reduce_sum(E)
```

```python
# Exercise: Find the variance and standard deviation of our 'E' tensor using Tensorflow Methods.

# find the variance of our tensor:
mean, variance = tf.nn.moments(E, axes=[0])
variance
```

```python
# standard deviation is the square root of the variance
# doesnt work with dtype int
std = tf.math.reduce_std(tf.cast(E, dtype=tf.float32))
std, std**2, variance
```

```python
# Other way to find variance: use tensorflow probability tfp
import tensorflow_probability as tfp
tfp.stats.variance(E)
```

```python
variance = tf.math.reduce_variance(tf.cast(E, dtype=tf.float32))
variance
```
